```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
def par(z1, z2):
    g = 1/z1 + 1/z2
    return 1/g

def ftow(f):
    return 2*np.pi*f
```

```python
r3 = 1E6
r2 = 10000
c1 = 4.7E-9
c2 = 1.1E-10
r4 = 10E3
r5 = 100000
r6 = 1000

def A1(f):
    w = ftow(f)
    zc = 1/(1j*w*c1)
    return -r3/(zc+r2)

def A2(f):
    w = ftow(f)
    zc = 1/(1j*w*c2)
    zo = par(zc, r5)
    return zo/r4

```

```python
x1 = np.linspace(100, 20000, 100000)
x2 = np.linspace(100, 100000, 100000)
plt.plot(x1, np.abs(A1(x1)))
plt.title("H(jw)")
plt.show()

plt.plot(x1, np.abs(A2(x1)))
plt.title("H(jw)")
plt.show()

plt.semilogx(x2, np.abs(A1(x2)*A2(x2)))
plt.title("Band Pass Transfer Function")
plt.xlabel("Frequency (Hz)")
plt.ylabel("H(jw)")
plt.savefig("lab2_bandpass.png",dpi=600)
plt.show()
```
